```python
import numpy as np
import pandas as pd
```

```python
# Load data
df = pd.read_csv("Superstore.csv", encoding='latin1')
```

```python
#null check
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y', dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y', dayfirst=True)
```

```python
# Remove outliers
q1 = df['Profit'].quantile(0.25)
q3 = df['Profit'].quantile(0.75)
```

```python
iqr = q3 - q1
df = df[(df['Profit'] >= (q1 - 1.5 * iqr)) & (df['Profit'] <= (q3 + 1.5 * iqr))]
```

```python
df_subset = df[df['Sales'] > 1000]
df_subset['Profit Margin'] = (df_subset['Profit'] / df_subset['Sales']) * 100
```

```python
df.loc[:, 'Profit Margin'] = (df['Profit'] / df['Sales']) * 100
```

```python
# Save cleaned data
df.to_csv('cleaned_superstore_sales.csv', index=False)
```
